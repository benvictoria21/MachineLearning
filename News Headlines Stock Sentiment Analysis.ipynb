```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('dataset/Combined_News_DJIA.csv', encoding = 'ISO-8859-1')


df
```

```python
df.head()
```

```python
df.shape
```

```python
train = df[df['Date']<'20150101']
test = df[df['Date'] > '20141231']
```

```python
train.shape
```

```python
test.shape
```

```python
data = train.iloc[:, 2:27]
data.replace("[^a-zA-Z]", " ", regex=True, inplace=True)
```

```python
list1 = [i for i in range(25)]
new_Index = [str(i) for i in list1]
data.columns = new_Index
data.head(5)
```

```python
for index in new_Index:
    data[index] = data[index].str.lower()
data.head(2)
```

```python
' '.join(str(x) for x in data.iloc[1, 0:25]) 
```

```python
headlines = []
for row in range(0, len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row, 0:25]))
headlines[5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
countVector = CountVectorizer(ngram_range=(2,2))
#ngram(2,2) means it will combine the 2 words together and assign the value

trainDataset = countVector.fit_transform(headlines)
trainDataset[0] 
```

```python
randomClassifier = RandomForestClassifier(n_estimators=200, criterion='entropy')
randomClassifier.fit(trainDataset, train['Label'])
```

```python
testTransform =[]
for row in range(0, len(test.index)):
    testTransform.append(' '.join(str(x) for x in test.iloc[row, 2:27]))
    
test_dataset = countVector.transform(testTransform)
predictions = randomClassifier.predict(test_dataset)
test.loc[1611, :]
```

```python
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
matrix = confusion_matrix(test['Label'], predictions)
print(matrix)
score = accuracy_score(test['Label'], predictions)
print(score)
report = classification_report(test['Label'], predictions)
print(report)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
# implememt BAG OF WORDS
tfidfVector = TfidfVectorizer(ngram_range=(2,2))
#ngram(2,2) means it will combine the 2 words together and assign the value

trainDataset = tfidfVector.fit_transform(headlines)
randomClassifier = RandomForestClassifier(n_estimators=200, criterion='entropy')
randomClassifier.fit(trainDataset, train['Label'])
```

```python
testTransform =[]
for row in range(0, len(test.index)):
    testTransform.append(' '.join(str(x) for x in test.iloc[row, 2:27]))
    
test_dataset = tfidfVector.transform(testTransform)
predictions = randomClassifier.predict(test_dataset)
predictions
```

```python
matrix = confusion_matrix(test['Label'], predictions)
print(matrix)
score = accuracy_score(test['Label'], predictions)
print(score)
report = classification_report(test['Label'], predictions)
print(report)
```

```python
from sklearn.naive_bayes import MultinomialNB
naive = MultinomialNB()
naive.fit(trainDataset, train['Label'])
```

```python
predictions = naive.predict(test_dataset)
predictions
```

```python
matrix = confusion_matrix(test['Label'], predictions)
print(matrix)
score = accuracy_score(test['Label'], predictions)
print(score)
report = classification_report(test['Label'], predictions)
print(report)
```
