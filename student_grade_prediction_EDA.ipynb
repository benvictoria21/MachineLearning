```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/benvictoria17/MachineLearning/master/dataset/student-mat.csv')
df.head()
```

```python
l=['address', 'famsize', 'Pstatus', 'Medu', 'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian', 'paid', 'nursery', 'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc']
df.drop(l, axis=1, inplace=True)
df.head()
```

```python
df.school.unique()
```

```python
sch = df.school.value_counts().to_dict()
plt.figure(figsize=(10,10))
plt.pie(x=sch.values(), labels=sch.keys(), autopct='%1.1f%%')
plt.show()
```
